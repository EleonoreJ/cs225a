```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw2/controller1.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t = []
    t7 = []
    for row in readCSV:
        t.append(float(row[0]))
        t7.append(float(row[1]))
```

```python
t = np.array(t)
t7 = np.array(t7)
plt.plot(t, t7)
plt.title("Joint 7 trajectory")
plt.xlabel("time")
plt.ylabel("theta7")
plt.show()
#print(t7[220],t7[-150])
print("frequency:" + str(2*np.pi*18/(t[-150] - t[220])) + " Hz")
```
