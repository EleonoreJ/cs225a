```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw1/controller5.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1_5 = np.array(t1)
plt.plot(t1_4, label= "q4")
plt.plot(t1_5, label= "q5")
plt.legend()
plt.title("Joint 1 trajectory")
plt.show()
t3_5 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3_4, label= "q4")
plt.plot(t3_5, label= "q5")
plt.legend()
plt.show()
t4_5 = np.array(t4)
plt.plot(t4_4, label= "q4")
plt.plot(t4_5, label= "q5")
plt.title("Joint 4 trajectory")
plt.legend()
plt.show()
```

I kept kv = 40. We changed the mass in the simulation but not in the controller. Therefore, the estimates of the dynamics is wrong, which leads to disturbances on the trajectories of the joints. In addition, the gravity vector is wrong, this is why joint 4 does not reach the desired position. Disturbances are allow due a wrong estimate of the centrigugal and coriolis forces.
