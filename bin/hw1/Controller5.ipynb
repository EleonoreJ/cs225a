```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw1/controller5.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1 = np.array(t1)
plt.plot(t1)
plt.title("Joint 1 trajectory")
plt.show()
t3 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3)
plt.show()
t4 = np.array(t4)
plt.plot(t4)
plt.title("Joint 4 trajectory")
plt.show()
```

```python
with open('../../hw1/controller5.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1_5 = np.array(t1)
plt.plot(t1_4, label= "q4")
plt.plot(t1_5, label= "q5")
plt.legend()
plt.title("Joint 1 trajectory")
plt.show()
t3_5 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3_4, label= "q4")
plt.plot(t3_5, label= "q5")
plt.legend()
plt.show()
t4_5 = np.array(t4)
plt.plot(t3_4, label= "q4")
plt.plot(t3_5, label= "q5")
plt.title("Joint 4 trajectory")
plt.legend()
plt.show()
```
