```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw1/controller3.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1 = np.array(t1)
plt.plot(t1)
plt.title("Joint 1 trajectory")
plt.show()
t3 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3)
plt.show()
t4 = np.array(t4)
plt.plot(t4)
plt.title("Joint 4 trajectory")
plt.show()
```

```python
with open('../../hw1/controller3.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1_3 = np.array(t1)
plt.plot(t1_2, label= "q2")
plt.plot(t1_3, label= "q3")
plt.legend()
plt.title("Joint 1 trajectory")
plt.show()
t3_3 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3_2, label= "q2")
plt.plot(t3_3, label= "q3")
plt.legend()
plt.show()
t4_3 = np.array(t4)
plt.plot(t4_2, label= "q2")
plt.plot(t4_3, label= "q3")
plt.title("Joint 4 trajectory")
plt.legend()
plt.show()
```
