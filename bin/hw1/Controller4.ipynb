```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw1/controller4.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1 = np.array(t1)
plt.plot(t1)
plt.title("Joint 1 trajectory")
plt.show()
t3 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3)
plt.show()
t4 = np.array(t4)
plt.plot(t4)
plt.title("Joint 4 trajectory")
plt.show()
```

```python
with open('../../hw1/controller4.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
t1_3 = np.array(t1)
plt.plot(t1_2, label= "q3")
plt.plot(t1_3, label= "q4")
plt.legend()
plt.title("Joint 1 trajectory")
plt.show()
t3_3 = np.array(t3)
plt.title("Joint 3 trajectory")
plt.plot(t3_2, label= "q3")
plt.plot(t3_3, label= "q4")
plt.legend()
plt.show()
t4_3 = np.array(t4)
plt.plot(t4_2, label= "q3")
plt.plot(t4_3, label= "q4")
plt.title("Joint 4 trajectory")
plt.legend()
plt.show()
```

The estimate of the dynamics, centrifugal and coriolis forces and gravity are very accurate which allows the trajectories to be greatly corrected. For this controller, we can have kv based on the dynamics of the system as we compensated the non linearities. 
$$ \ddot{e} + k_v \dot{e} + k_p e = 0 $$
By identification with $$ \ddot{e} + 2\xi w_n \dot{e} + w_n^2 e $$
Finally we have: $kv = 2 \xi \sqrt{kp} = 40$
