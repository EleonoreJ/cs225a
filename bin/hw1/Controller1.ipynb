```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw1/controller1.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t1 = []
    t3 = []
    t4 = []
    for row in readCSV:
        #print(row)
        t1.append(float(row[0]))
        t3.append(float(row[1]))
        t4.append(float(row[2]))
```

```python
t1 = np.array(t1)
plt.plot(t1)
plt.title("Joint 1 trajectory")
plt.show()
```

```python
t3 = np.array(t3)
plt.plot(t3)
plt.title("Joint 3 trajectory")
plt.show()
```

```python
t4 = np.array(t4)
plt.plot(t4)
plt.title("Joint 4 trajectory")
plt.show()
```
