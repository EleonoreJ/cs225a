# 2.d Joints limits

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller2.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    t4 = []
    t6 = []
    xdx = []
    xdy = []
    xdz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
        t4.append(float(row[6]))
        t6.append(float(row[7]))
plt.plot(t_posx, label="x")
plt.plot(t_posy, label="y")
plt.plot(t_posz, label="z")
plt.plot(xdx, label="x desired")
plt.plot(xdy, label="y desired")
plt.plot(xdz, label="z desired")
plt.title("End effector trajectories")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(t4, label = "Joint 4")
plt.plot(t6, label = "Joint 6")
plt.title("Joints trajectory")
plt.xlabel("iterations")
plt.ylabel("Joint angles (rad)")
plt.legend()
plt.show()
```

```python
print("minimum angle Joint 4:" + str(180*min(t4)/np.pi))
print("extrema angles Joint 6:" + str(180*min(t6)/np.pi) + "," + str(180*max(t6)/np.pi))
```

The system is able to track the desired positions, but Joint 4 reaches its limit which is -170. The minimum angle is -190. Joint 6 does not seem to reach its limits (0, 210).

Controller from 2.e and 2.f is preferable because the robot reaches the desired position and avoid as possible joints limits. Indeed, the top priority is to reach the desired position.
