```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller3.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    phix = []
    phiy = []
    phiz = []
    xdx = []
    xdy = []
    xdz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
        phix.append(float(row[6]))
        phiy.append(float(row[7]))
        phiz.append(float(row[8]))
plt.plot(t_posx, label="x")
plt.plot(t_posy, label="y")
plt.plot(t_posz, label="z")
plt.plot(xdx, label="x desired")
plt.plot(xdy, label="y desired")
plt.plot(xdz, label="z desired")
plt.title("End effector trajectories")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(phix, label = "delta phi x")
plt.plot(phiy, label = "delta phi y")
plt.plot(phiz, label = "delta phi z")
plt.title("Orientation errors")
plt.xlabel("iterations")
plt.ylabel("orientation error (rad)")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller3.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    xdx = []
    xdy = []
    xdz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
k = 2000
t = np.linspace(k, len(xdx), len(xdx)-k)
plt.plot(t,t_posx[k:], label="x")
plt.plot(t,xdx[k:], label="x desired")
plt.title("End effector Disturbances")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(t,t_posy[k:], label="y")
plt.plot(t,xdy[k:], label="y desired")
plt.title("End effector Disturbances")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(t,t_posz[k:], label="z")
plt.plot(t,xdz[k:], label="z desired")
plt.title("End effector Disturbances")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
```

We can see that the robot reaches the desired position and orientation. The desired position is achieved with critical damping. However, when the robot reaches the desired position, the orientation error starts to increase to finally converge. It is due to the fact that we might be close to a singularity. Therefore, the orientation error increases and then converges. However, when we zoom in we can see that changing the orientation disturbs the end-effector position.
