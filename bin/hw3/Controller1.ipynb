# PD Controller

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller1.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    xdx = []
    xdy = []
    xdz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
plt.plot(t_posx, label="x")
plt.plot(t_posy, label="y")
plt.plot(t_posz, label="z")
plt.plot(xdx, label="x desired")
plt.plot(xdy, label="y desired")
plt.plot(xdz, label="z desired")
plt.title("End effector trajectories")
plt.legend()
plt.show()
```

The actual trajectory matches perfectly for the z direction, but not for the x and y directions. It matches for the z direction because it is static. The differences for the sinusoids come from the fact that trajectories are dynamics. The model only takes into account position of the end effector and not the velocity, this is why it is late compared to the desired position
