# Velocity Analysis

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller4.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    xdx = []
    xdy = []
    xdz = []
    vx = []
    vy = []
    vz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
        vx.append(float(row[6]))
        vy.append(float(row[7]))
        vz.append(float(row[8]))
plt.plot(t_posx, label="x")
plt.plot(t_posy, label="y")
plt.plot(t_posz, label="z")
plt.plot(xdx, label="x desired")
plt.plot(xdy, label="y desired")
plt.plot(xdz, label="z desired")
plt.title("End effector trajectories")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(vx, label="x vel")
plt.plot(vy, label="y vel")
plt.plot(vz, label="z vel")
plt.title("Velocity")
plt.xlabel("iterations")
plt.ylabel("velocity (m/s)")
plt.legend()
plt.show()
```

The maximum velocity Vmax is equal to 0.1. On the plot, it is obvious that the robot velocity clearly exceeds the limit. Every component of the velocity exceeds 1m/s in magnitude. Therefore, it is essential to implement a saturation.

# Velocity saturation

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('../../hw3/controller4.txt') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    t_posx = []
    t_posy = []
    t_posz = []
    xdx = []
    xdy = []
    xdz = []
    vx = []
    vy = []
    vz = []
    for row in readCSV:
        t_posx.append(float(row[0]))
        t_posy.append(float(row[1]))
        t_posz.append(float(row[2]))
        xdx.append(float(row[3]))
        xdy.append(float(row[4]))
        xdz.append(float(row[5]))
        vx.append(float(row[6]))
        vy.append(float(row[7]))
        vz.append(float(row[8]))
plt.plot(t_posx, label="x")
plt.plot(t_posy, label="y")
plt.plot(t_posz, label="z")
plt.plot(xdx, label="x desired")
plt.plot(xdy, label="y desired")
plt.plot(xdz, label="z desired")
plt.title("End effector trajectories")
plt.xlabel("iterations")
plt.ylabel("Positions (m)")
plt.legend()
plt.show()
plt.plot(vx, label="x vel")
plt.plot(vy, label="y vel")
plt.plot(vz, label="z vel")
plt.title("Velocity")
plt.xlabel("iterations")
plt.ylabel("velocity (m/s)")
plt.legend()
plt.show()
```

We observe that the maximum velocity is not exceeded for each component of the velocity vector. It influences the response of the robot. Instead of the usual response, we can observe straight lines in the trajectory of the end-effector.
